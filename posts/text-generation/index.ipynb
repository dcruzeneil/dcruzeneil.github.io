```python
path = "lyrics.txt"
with open(path, "r") as file:
    lyrics = file.read()
```

#### <font color="green"> Tokenizing the Data </font>

```python
import nltk
import re
from nltk.tokenize import regexp_tokenize
```

```python
tokenPattern = r"\w+|[^\w\s]|\n"
```

```python
tokenizedLyrics = nltk.regexp_tokenize(lyrics, tokenPattern)
```
