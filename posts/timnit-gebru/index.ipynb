- Black Representation low in Computer Vision (homogeneous) → Computer Vision harmful for Black community
- For same technology, based on your experience you might have different understandings - what the pros and cons could be.
- Faceception → startup which predicts based on your facial structure your IQ, whether you are an offender, terrorist, and so on + this could lead to people in marginalized groups being labelled as “terrorists” → systematic racism
- HireVue (startup) → uses your face during interviews to detect your internal state (what you are feeling internally) → first of all, we do not know if this will adhere with the equal employment opportunities law because for instance if there is a glitch which always claims you are “angry” and if every company starts using this then you do not have a chance to get a job anywhere + furthermore, there is close to none representation of marginalized groups when it comes to propagating for the use of such software + just like the FDA approves a drug before it starts being used there should be a body which regulates and understands the use of algorithms before it is publicized and is being used publicly
- Maryland, Baltimore (face recognition used by police) → most invasive and extensive face recognition + they used their software to capture protestors and then link them to their socials + then later arrested them for unrelated reasons
