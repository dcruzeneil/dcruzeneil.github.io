# Dr. Timnit Gebru's Talk on Computer Vision 

In this talk, Dr. Timnit Gebru talks about the different nuances of the intersectionality of <i>race</i> and <i>computer vision</i>. She begins her argument by making it clear that computer vision <b>is</b> harming certain marginalized communities, for example the black community, and it is these very communities which have extremely low representation in the field, exacerbating the problem even further. She divides the ongoing problem with computer vision into two parts:
<ul>
    <li> Representation in the Data-Set
    <li> Representation beyond Data-Set
</ul>

## Representation in the Data-Set

Through numerous examples, Dr. Gebru highlights how certain applications of computer vision have questionable impacts on people of certain communities. She starts off by talking about <b>FACECEPTION</b> a startup which predicts your "personality" or "group" based on your facial structure. After visiting their website, I found out that their "classification" could be something as trivial as "Professional Poker Player" to something as serious as "terrorist". This system of recognizing "terrorists" based on facial structure, will ignore everything else, including but not limited to government-funded terrorism, and will amplify systematic racism - with certain marginalized groups being more likely to be labelled as "terrorists". She further talks about a startup called <b>HireVue</b> and links it to the <i>automation bias</i>. HireVue analyzes your face from during an interview to try to detect your "internal state". There is no proof if this method adheres to Equal Employment Opportunity standard, and the flaws of the system combined with automation bias - blindly trusting this system, when it is wrong - would lead to some people being constantly rejected from jobs.<br>

She talks about how a lot of these problems stem from datasets not being diverse. Furthermore, she highlights how this problem is not just specific to computer vision, but also manifests itself in areas like "air-bag" testing for cars, and drug-testing on humans to name a few. This is because a lot of the publicly available dataset is of lighter shaded men and have very few darker shaded women. Furthermore, through an example of a classifier which classifies whether a given picture is of a wedding or not, she highlights how the model's idea of a wedding was very heteronormative and western. Most of the times these computer vision applications are invasive, like the one used by Baltimore police, therefore the idea of having more diverse datasets is definitely an important one. But is that enough?

## Representation beyond Data-Set

Apart from the need of representation in the dataset, we must also talk about "structural representation" - having people who share your interests. The reason why the computer vision community needs structural representation is because there are systems being developed and employed, which have a negative impact on certain communities, and more often than not, these communities are not part of the decision making process. For example, there was a system being developed - which we also discussed in class - which assign or predict the gender of the person based on their facial structures. First of all, one can imagine how the very assumption that a system can predict someone's gender based on their facial structure is deep-rooted in stereotypical notions of "gender". Combine that with the fact that sometimes this face data is obtained unethically, and lastly combine that with question: why do we need a system which can predict people's gender? It is harmful and puts the transgender community in danger, especially in conservative countries, and the only people it is potentially helping is multi-national corporation, for their targeted advertising. And even this targeted advertising would further gender-stereotypes, as the very implication of gender-targeted advertising is that you have a box of things that you think people of a specific gender do and enjoy. She also talks about imperfect systems combined with imperfect use aggravates bias. She talked about how not too long ago, people were not expected to disclose how a certain model worked and what was the error rate. And furthermore, sometimes people did not even use a model (which is imperfect to begin with) the way it was intended to be used. For instance, people would use sketches and photos of look-alike actors/actresses in models which explicitly need the photo of the suspect and this would lead to imperfect accusations, or at times arrests. <br>

One thing which really stuck out to me was the parallel she made from during the WW-II, where she presented a newspaper article which highlighted how to tell apart "Japanese Americans" from "Chinese Americans". She linked this to our current applications of computer vision, and talked about if this is really the direction we want to go in. 

## Too Long; Didn't Read (TL;DR) 
The field of computer vision lacks both structural representation and representation in datasets, which puts certain communities at risk of negatively being affected by the applications of computer vision. Furthermore, these are the communities which are not involved in any of the decision-making processes which further exacerbates the issue.
