## What is a Neuron?

Let us think of the neuron like a box. There are numerical <b>inputs</b> coming into the neuron, and each of these inputs have associated numerical <b>weights</b>. That is, there is a weight associated with each input. When the neuron receives these inputs and their respective weights, it does what is called a <b>weighted sum</b>. This means that the neuron takes each input and multiplies it with its associated weight, and then adds all of these values together. But that is not all, the neuron has a numerical value within it called a <b>bias</b> and the neuron adds this bias to the weighted sum.

So altogether, a neuron does the following: 
<ol>
    <li> Multiply each input with its associated weight
    <li> Add all these values together
    <li> Add the bias to the value from above
</ol>

```python
def neuron(inputs, weights):
    bias = 0.5
    output = inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2] * weights[2] + bias
    return output
    
inputs = [0.3, 0.6, 0.9]
weights = [0.1, 0.2, 0.5]

print(neuron(inputs, weights))
```

Therefore in the above code block we can see a neuron which takes in three inputs multiplies each of those inputs with their associated weight, and then adds the bias. So there were three inputs, going into one neuron. Now let us imagine three inputs going into three neurons. This would mean a total of $3 * 3 = 9$ weights because all three inputs have three different lines going into each of the three neurons, and each of these lines have different weights associated with thim.
